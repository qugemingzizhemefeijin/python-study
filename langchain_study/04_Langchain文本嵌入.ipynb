### 什么是文本嵌入

- 文本嵌入（Text Embedding）是将文本（如单词、句子和段落）转换成固定长度的数值向量表示。
- 这些向量捕捉文本的语义信息，相似语义的文本在向量空间中距离较近

例如：
- "Hi there!"和"Oh, Hello!"的向量相似度高？（都是打招呼）
- "What's your name?"和"My Friends call me world"的向量相似度较低（一个在问问题，一个在回答问题）

### 为什么需要文本嵌入

电脑不认识文字，对自然语言理解的不够。但它能算数字！通过这些向量，电脑就能理解文字的含义。

### 文本嵌入能做什么？

- 找相似内容：让电脑明白哪些文字意思相近，比如搜索的时候不只是找关键词，还能找到意思差不多的内容。
- 分类文字：比如看一条评论是好评还是差评，电脑通过向量就能判断。
- 推荐东西：像推荐文章，电影啥的，基于内容的意思来推荐，而不是瞎猜。
- 跨语言也能搞：中文、英文都能转成向量，意思相近的还能凑一块。

```python
# 智普AI
import os
from langchain_community.chat_models import ChatZhipuAI
from langchain_core.messages import AIMessage, HumanMessage, SystemMessage
import warnings

warnings.filterwarnings("ignore", message=".*HMAC key is 16 bytes long.*")

os.environ["ZHIPUAI_API_KEY"] = os.getenv('ZHIPU_API_KEY')

chat = ChatZhipuAI(
    model="glm-4",
    temperature=0.5,
)
```

```python
from langchain_community.embeddings import ZhipuAIEmbeddings

# 加载模型
embeddings = ZhipuAIEmbeddings(model="embedding-2")

# 准备评论
comments = ["这个产品太棒了！", "一天就坏了，垃圾！"]
# 转换成向量
vectors = embeddings.embed_documents(comments)

print(f"向量的维度：{len(vectors[0])}")
print(f"向量的维度：{vectors[0]}")
```

```python
test_comment = ['这个东西超好用！']
test_vector = embeddings.embed_documents(test_comment)
print(f"向量的维度：{test_vector[0]}")

from sklearn.linear_model import LogisticRegression

labels = [1, 0] # 1好评，0差评
clf = LogisticRegression().fit(vectors, labels)

# 预测
prediction = clf.predict(test_vector)
print("好评" if prediction[0] == 1 else "差评")
```

### 练习

```python
# 导入模块
from langchain_community.document_loaders import TextLoader # 加载文档
from langchain.text_splitter import RecursiveCharacterTextSplitter # 对文本进行拆分

# 加载文件
loader = TextLoader("./政策文件.txt", encoding='utf-8')
documents = loader.load()

# 中文分块
text_splitter = RecursiveCharacterTextSplitter(
    chunk_size=50, # 适合中文段落（段落大小）
    chunk_overlap=0, # 避免上下文缺失(保持段落完整性)
    separators=["\n\n", "\n", "。", "！", "？"],
)
splits = text_splitter.split_documents(documents)
print(f'生成{len(splits)}个分块')
print(f'第1块内容：{splits[0].page_content}')
```

```python
# 本地模型
from langchain_huggingface import HuggingFaceEmbeddings

# 加载本地模型
embeddings = HuggingFaceEmbeddings(
    model_name='tmp/bge-small-zh-v1.5',
    model_kwargs={'device':'cpu'},
    encode_kwargs={'normalize_embeddings': False} # 提升相似度计算精度
)
```

```python
import os
from langchain_community.chat_models import ChatZhipuAI
from langchain_core.messages import AIMessage, HumanMessage, SystemMessage
import warnings

warnings.filterwarnings("ignore", message=".*HMAC key is 16 bytes long.*")

os.environ["ZHIPUAI_API_KEY"] = os.getenv('ZHIPU_API_KEY')

chat = ChatZhipuAI(
    model="glm-4",
    temperature=0.5,
)

# 网络版模型
from langchain_community.embeddings import ZhipuAIEmbeddings

# 加载模型
embeddings = ZhipuAIEmbeddings(model="embedding-2")
```

```python
# pip install chromadb # 持久化工具
# 数据持久化(将内容向量化支持存储到数据库中)
from langchain_community.vectorstores import Chroma

db = Chroma.from_documents(
    documents=splits,               # 输入的文本
    embedding=embeddings,           # 选择模型
    persist_directory="./tmp/chroma_db" # 数据库存储的目录
)

# Collection expecting embedding with dimension of 512, got 1024
# 这个是因为向量维度不匹配，再次运行会存在chroma_db的数据文件，会造成不匹配

# 语义搜索
docs = db.similarity_search("迟到的处罚", k=1) # k是匹配的内容几段
print('匹配结果：')
for doc in docs:
    print(doc.page_content)
```
