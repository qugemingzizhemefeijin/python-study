```python
import os

from langchain_community.llms import tongyi

key = os.getenv('TONGYI_API_KEY')
llm = tongyi.Tongyi(api_key=key)

# 调用模型使用
info = llm.invoke("夏天适合吃什么水果？回复几个水果名称就可以来！")
print(info)
```

```python
# 智普AI

import os
from langchain_community.chat_models import ChatZhipuAI
from langchain_core.messages import AIMessage, HumanMessage, SystemMessage
import warnings

# 忽略警告
# 智谱 AI 的 API 密钥长度是 16 字节（32 个十六进制字符），而使用 SHA256 算法的 HMAC 签名，RFC 7518 建议密钥至少 32 字节（256 位）以保证安全性。
# 这个警告是来自 python-jose 或 jwt 库的安全提示，通常不影响代码运行，但表明你的 HMAC 密钥长度不足 32 字节。
# 此处可以忽略掉那个错误。
warnings.filterwarnings("ignore", message=".*HMAC key is 16 bytes long.*")

key = os.getenv('ZHIPU_API_KEY')

os.environ["ZHIPUAI_API_KEY"] = key

chat = ChatZhipuAI(
    model="glm-4",
    temperature=0.5,    # 模型温度，0-1之间。 值越小，随机性越低。
)

messages = [
    AIMessage(content="Hi."),   # 这个是有点像网页版本的一打开界面，AI自动说一句：有什么我可以帮你得吗？
    SystemMessage(content="你是一个水果品鉴大师，同时中文水平也非常的厉害。"), # 模型的角色
    HumanMessage(content="夏天适合吃什么水果？"), # 提问
]

response = chat.invoke(messages)
print(response.content)

```
